```python
#12. Remove the un-necesary column from dataframeand storing to new DF : customer_df2
customer_df2 = customer_df1.drop(columns = 'Not_Useful_Column')
customer_df2
```

```python
#13. Removing non-aplhabets  from column values= ' LAST NAME'
#strip() is used to remove all the white spaces by default
#strip('/') : used to remove / from both front and back
#lstrip('/') : usedto remvoe / from left of the string
# Example row :
#11	1012	Harry	...Potter	
#14	1015	Toby	Flenderson_

###customer_df2['Last_Name']= customer_df2['Last_Name'].str.lstrip('/') 
###customer_df2['Last_Name']= customer_df2['Last_Name'].str.lstrip('...') 
###customer_df2['Last_Name']= customer_df2['Last_Name'].str.rstrip('_')

###

import re

# Remove all non-alphabetic characters (including slashes, dots, underscores, etc.)
customer_df2['Last_Name'] = customer_df2['Last_Name'].str.replace(r'[^a-zA-Z]', '', regex=True)
customer_df2
```

```python
#14 : Standardise the phone number column format  example issues below: 
#123-545-5421
#123/643/9775
# 7066950392
#123-543-2345
customer_df2['Phone_Number'] = customer_df2['Phone_Number'].str.replace(r'[^a-zA-Z0-9]', '', regex=True)
customer_df2['Phone_Number']

```

```python
#15 : Split the address column
#customer_df2['Address'].str.split(',',n=1)
customer_df2['Address'].str.split(',',n=2,expand=True)
customer_df2[['Street_Name','State','Zip_Code' ]]= customer_df2['Address'].str.split(',',n=2,expand=True)
customer_df2
```

```python
#16: inconsisted data of YES, Y in payin customer & do not contact cols

customer_df2['Paying Customer'] = customer_df2['Paying Customer'].replace({'Y':'Yes','N':'No'},regex = False)
customer_df2
```

```python
#17: inconsisted data of YES, Y in do not contact cols

customer_df2['Do_Not_Contact'] = customer_df2['Do_Not_Contact'].replace ({'Y':"Yes",'N':'No'},regex = False)
customer_df2
```

```python
# get list of customers who didnt have the phone number :
import re
customer_df2['Phone_Number']=customer_df2['Phone_Number'].str.replace('Na',"")
# Find rows with NaN or empty values
customers_without_phone_df4= customer_df2[(customer_df2['Phone_Number'] == "") | (customer_df2['Phone_Number'].isna())]
customers_without_phone_df4
```

```python

row_index_without_phone = customer_df2[(customer_df2['Phone_Number'] =="")|(customer_df2['Phone_Number'].isna())].index

customers_withphone_df3 = customer_df2.drop(row_index_without_phone)
customers_withphone_df3 # this dataframe has customers who has phone numbers
```

```python
# reset the  customers dataframe index who are without phone numbers 




customers_without_phone_df4.reset_index()
```

```python
# reset the  customers dataframe index (customers_without_phone_df4) who are without phone numbers 


customers_without_phone_df4.reset_index(drop = True)
```

```python
#resetting the customer with phone numbers dataframe (customers_withphone_df3)

customers_withphone_df3.reset_index(drop = True)
```

```python
# now extract list of customer who have no issue to contact

customers_list_contact = customers_withphone_df3[customers_withphone_df3['Do_Not_Contact'] =='No']
customers_list_contact  # final list where we can contact customers and readyu  for analysis
```

```python
customers_list_contact=customers_list_contact.reset_index(drop = True)
customers_list_contact
```
